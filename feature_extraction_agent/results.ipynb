```python
import pickle
import os 

with open(os.path.join('evaluation_results', 'success'), 'rb') as f:
    success_rates = pickle.load(f)

with open(os.path.join('evaluation_results', 'true_labels'), 'rb') as f:
    true_labels = pickle.load(f)

with open(os.path.join('evaluation_results', 'pred_labels'), 'rb') as f:
    predicted_labels = pickle.load(f)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

for feature in true_labels.keys():
    unique_vals = list(set(true_labels[feature]))
    experience_cm = confusion_matrix(
        true_labels[feature],
        predicted_labels[feature],
        labels=unique_vals
        )

    plt.figure(figsize=(8, 6))
    sns.heatmap(experience_cm, annot=True, fmt='d', cmap='Blues', xticklabels=unique_vals, yticklabels=unique_vals)
    plt.title(f"Confusion Matrix for {feature}")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.show()

    # Accuracy for "experience_level"
    experience_accuracy = accuracy_score(
        true_labels[feature],
        predicted_labels[feature]
        )
    print(f"Accuracy for {feature}: {100*experience_accuracy:.2f}%")
```
